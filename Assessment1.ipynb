2.Composite Development Score:

Create a new metric called Composite Score Using formula:

Composite Score = 0.30 * life expectancy index + 0.30 * GNI per capita index

```python
sa_df['Composite Score'] = (
    0.30 * sa_df['life_expectancy'] +
    0.30 * sa_df['gross_inc_percap']
)
```

Rank South Asain countries based on Composite Score

```python
south_asia = ['Afghanistan', 'Bangladesh', 'Bhutan', 'India', 'Maldives', 'Nepal', 'Pakistan', 'Sri Lanka']
sa_df_current = df[df['country'].isin(south_asia)].copy()

sa_df_current['Composite Score'] = (
    0.30 * sa_df_current['life_expectancy'] +
    0.30 * sa_df_current['gross_inc_percap']
)

composite_ranking = (
    sa_df_current.groupby('country')['Composite Score'].mean()
    .reset_index()
    .sort_values(by='Composite Score', ascending=False).reset_index()
)
composite_ranking['rank'] = range(1, len(composite_ranking) + 1)
composite_ranking
```

Plot the top 5 countries in horizontal bar chart

```python
plt.figure()
top5 = composite_ranking.head(5)
plt.figure(figsize=(8,5))
plt.barh(top5['country'], top5['Composite Score'], color='lightpink')
plt.title('Top 5 South Asian Countries by Composite Score')
plt.xlabel('Composite Score')
plt.ylabel('Country')
plt.savefig('Top5_SouthAsia.png')
plt.show()
```

- Compare the ranking of countries by Composite Score with thier HDI ranking and discuss any diffrences.

```python
hdi_ranking = (
    sa_df_current.groupby('country')['hdi'].mean()
    .reset_index()
    .sort_values(by='hdi', ascending=False).reset_index()
)
hdi_ranking['HDI Rank'] = range(1, len(hdi_ranking) + 1)

comparison = composite_ranking.merge(hdi_ranking, on='country', how='left')
comparison[['country', 'Composite Score', 'rank', 'hdi', 'HDI Rank']]
```

3. Outliers Detection

Calculate IQR for HDI

```python
Q1_hdi = df_1b['hdi'].quantile(0.25)
Q3_hdi = df_1b['hdi'].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi

lower_hdi = Q1_hdi - 1.5 * IQR_hdi
upper_hdi = Q3_hdi + 1.5 * IQR_hdi
hdi_outliers = df_1b[(df_1b['hdi'] < lower_hdi) | (df_1b['hdi'] > upper_hdi)]
print(hdi_outliers[['country', 'hdi']])
```

Calculate IQR for GNI per capita

```python
Q1_gni = df_1b['gross_inc_percap'].quantile(0.25)
Q3_gni = df_1b['gross_inc_percap'].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni

lower_gni = Q1_gni - 1.5 * IQR_gni
upper_gni = Q3_gni + 1.5 * IQR_gni
gni_outliers = df_1b[(df_1b['gross_inc_percap'] < lower_gni) | (df_1b['gross_inc_percap'] > upper_gni)]
print(gni_outliers[['country', 'hdi']])
```

```python
outliers = df_1b[
    (df_1b['hdi'] < lower_hdi) |
    (df_1b['hdi'] > upper_hdi) |
    (df_1b['gross_inc_percap'] < lower_gni) |
    (df_1b['gross_inc_percap'] > upper_gni)
]
outliers[['country', 'year', 'hdi', 'gross_inc_percap']].head(10)
```

Scatter Plot GNI per capita vs HDI(Highlighting Outliers)

```python
plt.figure()
#No-outliers
sns.scatterplot(data=df_1b, x='gross_inc_percap', y='hdi')
#Outliers
sns.scatterplot(data=outliers, x='gross_inc_percap', y='hdi', color='red', marker='x')
plt.title('HDI vs Gross National Income')
plt.xlabel('Gross National Income (GNI) per Capita')
plt.ylabel('HDI')
plt.legend(['No Outliers', 'Outliers'])
plt.savefig('HDI_vs_GNI(Outliers).png')
plt.show()
```

7. Outliers Detection:
- Detect outliers in HDI {'hdi} and GNI per Capita{'gross_inc_percap'} for each region using the 1.5 * IQR rule

step1: define a function to detect outliers using IQR

```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound
```

step2: Detect outliers for each region

SOUTH ASIA

```python
sa_hdi_outliers, _, _ = detect_outliers_iqr(df_sa, 'hdi')
sa_gni_outliers, _, _ = detect_outliers_iqr(df_sa, 'gross_inc_percap')

print("South Asia HDI Outliers:")
print(sa_hdi_outliers[['country', 'year', 'hdi']])

print("\nSouth Asia GNI Outliers:")
print(sa_gni_outliers[['country', 'year', 'gross_inc_percap']])
```

Middle East

```python
me_hdi_outliers, _, _ = detect_outliers_iqr(df_me, 'hdi')
me_gni_outliers, _, _ = detect_outliers_iqr(df_me, 'gross_inc_percap')

print("Middle East HDI Outliers:")
print(me_hdi_outliers[['country', 'year', 'hdi']])

print("\nMiddle East GNI Outliers:")
print(me_gni_outliers[['country', 'year', 'gross_inc_percap']])
```
